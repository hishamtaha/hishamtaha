```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
```

```python
x = np.arange(0,50,2)
print(x)
```

```python
#1 Declate Figure
fig = go.Figure()

#2 Add Argument (Trace, Title, Update Axis)
fig.add_trace(go.Scatter(x=x,y=x**2))

#3 Show Graph
fig.show()
```

```python
n = 100
x = np.linspace(0,1,n)
y0 = np.random.randn(n) + 5
y1 = np.random.randn(n)
y2 = np.random.randn(n) - 5
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=x, 
                         y=y0,
                         mode = 'lines',
                         name = 'lines'))
fig.add_trace(go.Scatter(x=x, 
                         y=y1,
                         mode = 'lines+markers',
                         name = 'lines+markers'))
fig.add_trace(go.Scatter(x=x, 
                         y=y2,
                         mode = 'markers',
                         name = 'markers'))
fig.show()
```
