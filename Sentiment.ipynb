```python
scores= []
values = []

for tweet in df['text'].to_list():
  sentence = flair.data.Sentence(tweet)
  sentence_model.predict(sentence)

  scores.append(sentence.labels[0].score)
  values.append(sentence.labels[0].value)

df['scores'] = scores
df['values'] = values
```

```python
print('# of Positive Tweets : ',len(df[df['values']=='POSITIVE']))
print('# of Negative Tweets : ',len(df[df['values']=='NEGATIVE']))
```
